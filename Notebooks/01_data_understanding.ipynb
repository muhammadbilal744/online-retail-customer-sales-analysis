## Missing Values Assessment

The dataset contains missing values in two columns:

- Customer ID
- Description

Before applying any cleaning techniques, it is important to understand the potential business reasons behind these missing values.

Customer ID contains approximately 22.77% missing values.

These missing values may represent anonymous or unregistered customers rather than data entry errors. Further investigation is required before deciding whether to remove or retain these records.

The Description column contains a small number of missing values (0.41%).

The impact of these missing records is expected to be minimal; however, they should be examined during the data cleaning phase.

## Initial Data Quality Findings

Several data quality issues were identified during the initial inspection:

### Missing Customer IDs

Customer IDs are missing in a significant portion of the dataset.

Some transactions appear to be valid purchases despite the absence of customer identifiers, suggesting the possibility of anonymous or guest purchases.

### Negative Quantities

Negative quantities were observed in multiple records.

These records may represent product returns, cancellations, stock corrections, or other business processes rather than data entry errors.

### Zero Prices

Several transactions contain a unit price of zero.

These records require further investigation to determine whether they represent free items, adjustments, promotional transactions, or data quality issues.

### Missing Product Descriptions

A small number of records contain missing product descriptions and should be reviewed during the data cleaning phase.

```python
df[df["Quantity"] < 0].head(20)
```

Invoices starting with the letter "C" appear to represent cancelled or returned transactions.

These records contain negative quantities and may reflect product returns or order cancellations rather than standard sales transactions.

```python
cancelled_orders = df[
    df["Invoice"].astype(str).str.startswith("C")
]

print("Cancelled Records:", len(cancelled_orders))
```

```python
negative_qty = df[
    df["Quantity"] < 0
]

print("Negative Quantity Records:", len(negative_qty))
```

```python
( cancelled_orders.shape[0] == negative_qty.shape[0] )
```

```python
negative_not_cancelled = df[
    (df["Quantity"] < 0)
    &
    (~df["Invoice"].astype(str).str.startswith("C"))
]
negative_not_cancelled.head(20)
```

### Investigation Result

Not all negative quantity transactions are associated with invoices starting with "C".

This indicates that negative quantities may represent multiple business processes, including cancellations, returns, stock adjustments, or other operational activities.

Further investigation is required.

### Inventory Adjustment Records

A subset of negative quantity transactions does not correspond to cancelled invoices.

These records frequently exhibit the following characteristics:

- Negative quantities
- Zero unit prices
- Missing customer identifiers
- Operational descriptions such as "lost", "damages", and "short"

This suggests that these records may represent internal inventory adjustments rather than customer transactions.

```python
df[df["Price"] == 0][
    ["Invoice", "StockCode", "Description", "Quantity", "Price", "Customer ID"]
].head(30)
```

## Price Analysis Findings

Records with a unit price of zero appear to belong to multiple categories:

1. Inventory adjustments and stock corrections
2. Operational records (lost, damaged, short inventory)
3. Potential promotional or free products

Therefore, zero-priced records should not automatically be treated as data quality errors. Further business-context analysis is required before removing them.
