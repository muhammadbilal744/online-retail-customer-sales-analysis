## Zero and Negative Price Investigation

Records with zero or negative prices were investigated.

Findings:

- Many records represent damaged, lost, wet, missing, or manually adjusted inventory.
- These records do not represent actual customer purchases.
- They can distort revenue and product performance analysis.

Decision:
Remove records where Price <= 0.

```python
df_clean = df_clean[df_clean["Price"] > 0]
```

```python
# Remove duplicates
df = df.drop_duplicates()

# Remove inventory adjustment records
df = df[
    ~(
        df["Description"].isna() &
        df["Customer ID"].isna() &
        (df["Price"] == 0)
    )
]

# Remove cancellations
df = df[
    ~df["Invoice"].astype(str).str.startswith("C")
]

# Remove negative or zero quantities
df = df[df["Quantity"] > 0]

# Remove zero/negative prices
df = df[df["Price"] > 0]
```

```python
df["Revenue"] = df["Quantity"] * df["Price"]
```
